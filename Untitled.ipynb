## Textblob

- Es una biblioteca sencilla de Python que ofrece acceso a la API para diferentes tareas de PLN, como análisis de sentimientos, corrección ortográfica, etc.
- El analizador de sentimientos de Textblob devuelve dos propiedades para una oración de entrada:
    - La polaridad es un valor de punto flotante (float) que se encuentra entre [-1,1]; -1 indica sentimiento negativo y +1 indica sentimiento positivo.
    - La subjetividad también es un valor de punto flotante (float) que se encuentra en el rango [0,1]. Las oraciones subjetivas generalmente se refieren a opiniones, emociones o juicios personales.

## Opción 1: Usar VADER para el análisis de sentimientos en español

- VADER (Valence Aware Dictionary and sEntiment Reasoner) es un modelo diseñado para realizar análisis de sentimientos en texto corto y es efectivo con varios idiomas, incluido el español.
- VADER (Diccionario con reconocimiento de valencia para razonamiento de sentimientos) es otro popular analizador de sentimientos basado en reglas.
- Utiliza una lista de características léxicas (p. ej., palabras) etiquetadas como positivas o negativas según su orientación semántica para calcular el sentimiento del texto.
- Vader Sentimiento devuelve la probabilidad de que una oración de entrada dada sea: positiva, negativa y neutra.

Por ejemplo:

"¡La comida estuvo genial!"
Positiva: 99%
Negativa: 1%
Neutra: 0%

Estas tres probabilidades suman el 100%.

Veamos cómo usar VADER:

```python
!pip3 install vaderSentiment --break-system-packages
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
sentence = "The food was great!"
vs = analyzer.polarity_scores(sentence)
print("{:-<65} {}".format(sentence, str(vs)))
```

## Opción 2: Usar Transformers de Hugging Face

La librería Transformers de Hugging Face ofrece modelos preentrenados como BERT y DistilBERT que funcionan muy bien para tareas de análisis de sentimientos, y muchos de estos modelos han sido entrenados para trabajar en español.

En este caso, el modelo nlptown/bert-base-multilingual-uncased-sentiment es un modelo multilingüe que está entrenado para varios idiomas, incluido el español. El análisis devuelve una etiqueta de sentimiento y su probabilidad.

**Interpretación:**

El modelo devuelve una lista de diccionarios con la predicción, que normalmente incluirá una clase de sentimiento (por ejemplo, "POSITIVE", "NEGATIVE") y una puntuación asociada.

```python
!pip3 install transformers torch --break-system-packages
```

```python
from transformers import pipeline

# Usamos el pipeline de análisis de sentimientos con un modelo en español
sentiment_analyzer = pipeline("sentiment-analysis", model="nlptown/bert-base-multilingual-uncased-sentiment")

# Comentario en español
comentario = "Me encanta este producto, es increíble!"

# Realizamos el análisis de sentimientos
resultado = sentiment_analyzer(comentario)

# Mostrar el resultado
print(resultado)
```

El resultado que obtuviste de Transformers ([{'label': '5 stars', 'score': 0.9092629551887512}]) significa lo siguiente:

label: '5 stars': El modelo ha clasificado el comentario como muy positivo, similar a una calificación de "5 estrellas". Esto indica que el sentimiento general del comentario es claramente positivo.

score: 0.9092629551887512: La puntuación de 0.91 (cerca de 1) indica la confianza del modelo en esta clasificación. Cuanto más cerca esté el valor de 1, mayor es la certeza del modelo en que el sentimiento es positivo. En este caso, el modelo tiene un 91% de confianza en que el comentario es positivo (5 estrellas).

**¿Y si el score fuera negativo?**

Si el label fuera negativo (por ejemplo, algo como 'label': '1 star' o 'label': 'negative'), y el score fuera negativo, significaría que el modelo ha identificado el comentario como negativo.

```python
from transformers import pipeline

# Usamos el pipeline de análisis de sentimientos con un modelo en español
sentiment_analyzer = pipeline("sentiment-analysis", model="nlptown/bert-base-multilingual-uncased-sentiment")

# Comentario negativo
comentario_negativo = "Este producto es terrible, no sirve para nada."

# Realizamos el análisis de sentimientos
resultado = sentiment_analyzer(comentario_negativo)

# Mostrar el resultado
print(resultado)
```

# Mini-proyecto:

Automatizando un análisis de sentimientos!

```python
!pip3 install ace_tools --break-system-packages
```

```python
import pandas as pd
from transformers import pipeline

# Usamos el pipeline de análisis de sentimientos con un modelo en español
sentiment_analyzer = pipeline("sentiment-analysis", model="nlptown/bert-base-multilingual-uncased-sentiment")

# Lista de comentarios en español
comentarios = [
    "Este producto es increíble, me encanta.",
    "Me parece un producto promedio, nada fuera de lo común.",
    "No me gustó para nada, muy malo.",
    "Excelente calidad, totalmente recomendado.",
    "Horrible experiencia, nunca volveré a comprar.",
    "La entrega fue rápida, pero el servicio al cliente no fue excelente.",
    "Muy buen servicio, pero el producto podría mejorar.",
    "Es lo peor que he comprado, muy decepcionado.",
    "¡Me fascina! Sin duda volveré a comprar.",
    "No cumple con las expectativas, esperaba mucho más.",
    "Es un buen producto, aunque un poco caro.",
    "Muy satisfecho con la compra, todo perfecto.",
    "El producto está bien, pero no es lo que esperaba."
]

# Realizamos el análisis de sentimientos
resultados = []
for comentario in comentarios:
    resultado = sentiment_analyzer(comentario)[0]
    label = resultado['label']
    score = resultado['score']

    # Clasificación: Negativo, Neutral o Positivo
    if label == '1 star' or label == '2 stars':
        sentimiento = 'Negativo'
    elif label == '3 stars':
        sentimiento = 'Neutral'
    else:  # 4 stars o 5 stars
        sentimiento = 'Positivo'

    resultados.append({'Comentario': comentario, 'Calificación': label, 'Confianza': score, 'Sentimiento': sentimiento})

# Convertir los resultados en un DataFrame
df = pd.DataFrame(resultados)
```

```python
# Mostrar el DataFrame
df
```

```python
!pip3 install wordcloud --break-system-packages
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
# Filtrar comentarios según el sentimiento
comentarios_positivos = ' '.join(df[df['Sentimiento'] == 'Positivo']['Comentario'])
comentarios_negativos = ' '.join(df[df['Sentimiento'] == 'Negativo']['Comentario'])
comentarios_neutrales = ' '.join(df[df['Sentimiento'] == 'Neutral']['Comentario'])

# Generar las nubes de palabras
wordcloud_pos = WordCloud(width=800, height=400, background_color='white').generate(comentarios_positivos)
wordcloud_neg = WordCloud(width=800, height=400, background_color='white').generate(comentarios_negativos)
wordcloud_neu = WordCloud(width=800, height=400, background_color='white').generate(comentarios_neutrales)

# Mostrar las nubes de palabras
plt.figure(figsize=(20, 6))

# Positivo
plt.subplot(1, 3, 1)
plt.imshow(wordcloud_pos, interpolation='bilinear')
plt.title("Nube de Palabras - Sentimiento Positivo", fontsize=16)
plt.axis('off')

# Negativo
plt.subplot(1, 3, 2)
plt.imshow(wordcloud_neg, interpolation='bilinear')
plt.title("Nube de Palabras - Sentimiento Negativo", fontsize=16)
plt.axis('off')

# Neutral
plt.subplot(1, 3, 3)
plt.imshow(wordcloud_neu, interpolation='bilinear')
plt.title("Nube de Palabras - Sentimiento Neutral", fontsize=16)
plt.axis('off')

# Mostrar la figura
plt.tight_layout()
plt.show()
```
